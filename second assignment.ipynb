In this section I'm Implementing a linear search algorithm.

```python
def search(arr, n, x): 
    for i in range (0, n): 
        if (arr[i] == x): 
            return i; 
    return -1; 

arr = ['a', 'b' , 'c' , 'd' ]; 
x = 'f'; 
n = len(arr); 
result = search(arr, n, x) 
if(result == -1): 
    print("Element is not present in array") 
else: 
    print("Element is present at index", result); 
```

```python
def callback(x):
    result = search(lines, len(lines), x)
    if(result == -1):
        print("Element doesn't exist linear")
    else:
        print("Element is in the index: ", result)
```

In this section I'm Implementing a binary search algorithm.

```python
def binarySearch(arr, l, r, x): 
    while l <= r: 
        mid = l + (r - l) // 2; 
        if arr[mid] == x: 
            return mid 
        elif arr[mid] < x: 
            l = mid + 1
        else: 
            r = mid - 1 
    return -1
```

```python
def binary_callback(x):
    result = binarySearch(lines, 0, len(lines)-1, x)
    if(result == -1):
        print("Element doesn't exist")
    else:
        print("Element is in the index: ", result)
```
